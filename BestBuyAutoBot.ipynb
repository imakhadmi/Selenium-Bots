```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import chromedriver_autoinstaller
import random
import gc
from IPython.display import clear_output
from IPython.display import Audio, display


# Set browser driver options
from selenium.webdriver.chrome.options import Options
chrome_options = Options()
# chrome_options.add_argument("--incognito")
# chrome_options.add_argument("--headless")
# chrome_options.add_argument("--disable-gpu")
# chrome_options.add_argument("window-size=1024,768")

# Check and update chromedriver 
chromedriver_autoinstaller.install()

# Set path for browser driver
PATH = "folder/path/of/your/chromedriver"
# Apply options for browser driver
driver = webdriver.Chrome(PATH, options = chrome_options)

def BB_Auto_Buy():
    # Open browser at BestBuy product web page
    driver.get("https://www.bestbuy.com/product-of-your-interest")
    print(driver.title)

    # While the product has not been added to the cart
    while True:
        # Run garbage collector
        gc.collect()
        
        # Try adding the product to the cart
        if ("Add to Cart" in driver.page_source):
            try:
                element = WebDriverWait(driver, 1.5 ).until(
                    EC.presence_of_element_located((By.CLASS_NAME, "fulfillment-add-to-cart-button"))
                )
                element.click()
                time.sleep(3)
                # If the product is not added to the cart and the page requires a wait to add the product to the cart, try adding to the cart every 5 seconds
                while ("This item is not in your cart yet." in driver.page_source):# or ("This item is currently unavailable for online purchase." in driver.page_source) or ("There was a problem adding your product to cart." in driver.page_source):
                    clear_output()
                    print(f"Item not added yet. {time.strftime('%I:%M:%S %p, %B %d %Y', time.localtime())}")
                    print("Waiting for the 'Add to Cart' button to come back. Checking again in .1 seconds.")
                    element = WebDriverWait(driver, 10).until(
                        EC.presence_of_element_located((By.CLASS_NAME, "fulfillment-add-to-cart-button"))
                    )
                    element.click()
                    time.sleep(random.uniform(0.1,.3))
                    if driver.find_element_by_xpath("/html/body/div[3]/main/div[2]/div[3]/div[2]/div/div/div[7]/div[1]/div/div/div/button") and driver.find_element_by_xpath("/html/body/div[3]/main/div[2]/div[3]/div[2]/div/div/div[7]/div[1]/div/div/div/button").text == "Sold Out":
                        continue

                if ("This item is currently unavailable for online purchase." in driver.page_source) or ("There was a problem adding your product to cart." in driver.page_source):
                    continue

                # If print function is reached, the product has been successfulyy added to the cart, so break the while loop.
                print(f"Add to cart successful! {time.strftime('%I:%M:%S %p, %B %d %Y', time.localtime())}")
                break

            # If the "Add to Cart" button is not available, refresh the page in 10 seconds and try again
            except:
                print("Could not select add to cart button.")

        else:
            clear_output()
            print(f"Not in stock. {time.strftime('%I:%M:%S %p, %B %d %Y', time.localtime())}")
        # clear all cookies in scope of session
        driver.delete_all_cookies()
        time.sleep(random.uniform(0.1,0.5))
        driver.refresh()

    # Proceed to cart for checkout
    driver.get("https://www.bestbuy.com/cart")
    print(driver.title)
    
    # Select "Checkout" button
    checkoutBtn = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.CLASS_NAME, "checkout-buttons__checkout"))
            )
    checkoutBtn.click()
    
    time.sleep(random.uniform(2,2.5))
    
    # Enter login information
    driver.find_element_by_id("fld-e").send_keys("email@email.com")
    driver.find_element_by_id("fld-p1").send_keys("password")
    driver.find_element_by_css_selector("div[class='cia-form__controls ']").click()
    
    time.sleep(random.uniform(2,2.5))
    
    # If web page requires account verification
    if ("Verify Your Account" in driver.page_source):
        element = WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.ID, "smsDigits"))
                )
        element.send_keys("1234")
        driver.find_element_by_css_selector("div[class='cia-form__controls ']").click()
    
    # If web page requires another log in based on the items in the cart, enter the password info for the account again
    if ("You have items in your order that require you to sign in." in driver.page_source):
        driver.find_element_by_id("fld-p1").send_keys("password")
        driver.find_element_by_css_selector("div[class='cia-form__controls ']").click()
        time.sleep(3)
        
    # Enter on-file credit card security code and select "Place Order" button
    print(f"At the checkout menu... {time.strftime('%I:%M:%S %p, %B %d %Y', time.localtime())}")
    driver.find_element_by_id("credit-card-cvv").send_keys("000")
    driver.find_element_by_css_selector("div[class='button--place-order-fast-track']").click()
    print(f"ORDER PLACED! {time.strftime('%I:%M:%S %p, %B %d %Y', time.localtime())}")

def allDone():
    display(Audio(url="https://ccrma.stanford.edu/workshops/mir2014/audio/T37-vibraphone-8k.wav", autoplay=True))
    ## Insert whatever audio file you want above
```
